## Load the dataset
- Dateset: Boston Pricing(available in sklearn)

```python
'''
loading the dataset
'''

from sklearn.datasets import load_boston
boston = load_boston()
```

## Single-dimensional Regression

```python
'''
plot the trend of prices
'''

from matplotlib import pyplot as plt
plt.scatter(boston.data[:, 5], boston.target, color='r')
```

```python
'''
use the most basic regression: LinearRegression
'''

from sklearn.linear_model import LinearRegression
lr = LinearRegression()

#now split data and train on it
import numpy as np
x = boston.data[:, 5]
y = boston.target
x = np.transpose(np.atleast_2d(x))
lr.fit(x, y)
y_predicted = lr.predict(x)
```

```python
'''
plotting the graph
'''

fig, pred = plt.subplots()
pred.plot(boston.data[:, 5], y_predicted, color='blue')
pred.scatter(x, y)
plt.show()
```

```python
'''
measuring how good the fit is quantitatively,
using Root Mean Square Error(RMSE)
'''
from sklearn.metrics import mean_squared_error
# mse = mean_squared_error(y, lr.predict(x))
mse = mean_squared_error(y, y_predicted)
print("Mean squared error (of training data): {:.3}".format(mse))

# A more nicely interpretable value is the root of mean squared error
rmse = np.sqrt(mse)
print("RMSE (of training data): {:.3}".format(rmse))
# Note: One advantage of using RMSE is that we can quickly obtain a very rough estimate of
#       the error by multiplying it by two.
```

```python
'''
the above measure is still hard to comprehend,
so let us switch to a more relativistic metric,
the "r2_score metric".
'''
from sklearn.metrics import r2_score
r2 = r2_score(y, y_predicted)
print("R2 (on training data): {:.2}".format(r2))

# also we can use the in-built score function of the regressor
r2 = lr.score(x, y)
print("R2 (using the score method): {:.2}".format(r2))
```

## Multi-dimensional Regression

```python
'''
Lets try predicting a single output using multiple inputs.
--> MULTI-DIMENSIONAL REGRESSION
'''

x = boston.data
y = boston.target
lr.fit(x,y)
r2 = lr.score(x, y)
print("R2 (using the score method): {:.2}".format(r2))
```

```python
'''
let us now plot the prediction versus actual value.
'''

p = lr.predict(x)
plt.scatter(p, y)
plt.xlabel('Predicted price')
plt.ylabel('Actual price')
plt.plot([y.min(), y.max()], [[y.min()], [y.max()]], color='red')
plt.show()
```

## Cross-Validation for Regression
- To ensure that the regression holds true generally and not just for the training data, we must follow cross-validation.

- We will follow the KFold class to build a 5 fold cross validation and test the **generalization ability** of **linear regression** 

```python
'''
use the kFold class of sklearn, for cross-validation
'''
from sklearn.model_selection import KFold
kf = KFold(n_splits=5)
p = np.zeros_like(y)
# TODO: write an explanation of what the above function does
for train, test in kf.split(x):
    lr.fit(x[train], y[train])
    p[test] = lr.predict(x[test])
    
# the code below uses 'rmse' to find the error for the above cross-validation method used
rmse_cv = np.sqrt(mean_squared_error(p, y))
print ('RMSE on 5-fold CV: {:.2}'.format(rmse_cv))
```

#### IMPORTANT NOTES:
- Ordinary least squares is fast at learning and returns a simple model, which is fast at prediction time.
- For the above reason it should be the first model that we try in a regression problem.

## Penalized or regularized regression
- In ordinary regression, the returned fit is the best fit on the training data. This can lead to overfitting.
- This is where penalizing comes to our aid.
- Penalizing means that we add a penalty for *overconfidence* in the parameter values. Thus, we accept a 'slightly' worse fit in order to have a simpler model
- Three ways are:
    1. L1: Lasso
    2. L2: Ridge
    3. Combination of both: ElasticNet

## Using ElasticNet in scikit-learn

```python
from sklearn.linear_model import ElasticNet, Lasso
en = ElasticNet(alpha=0.5, l1_ratio=0.7)

en.fit(x,y)
mse = mean_squared_error(y, en.predict(x))
print("Mean squared error (of training data): {:.3}".format(mse))

# A more nicely interpretable value is the root of mean squared error
rmse = np.sqrt(mse)
print("RMSE (of training data): {:.3}".format(rmse))

# now we may use it like linear regression
for train, test in kf.split(x):
    lr.fit(x[train], y[train])
    p[test] = en.predict(x[test])
    
# the code below uses 'rmse' to find the error for the above cross-validation method used
rmse_cv = np.sqrt(mean_squared_error(p, y))
print ('RMSE on 5-fold CV: {:.2}'.format(rmse_cv))
```

### Conclusion:
   - The difference in error for training data is positively meager
    (i.e. 4.8 to 4.97)
   - The error for cross-validation falls considerably from 6.1 to
     5.0 

## Visualizing the Lasso Path

```python
# lasso has already been imported above
las = Lasso(normalize=1)
alphas = np.logspace(-5, 2, 1000)
alphas, coefs, _= las.path(x, y, alphas=alphas)
'''
-> So what does the above code do?
----> For each value in aplhas, the path method on the Lasso object
      returns the coefficientsthat solve the lasso problem with the parameter
      value.
'''

fig, ax = plt.subplots()
ax.plot(alphas, coefs.T)
# Set log scale
ax.set_xscale('log')
# Make alpha decrease from left to right
ax.set_xlim(alphas.max(), alphas.min())
plt.xlabel('Aplha')
plt.ylabel('Coefficient of weight')
plt.show()
```

### Conclusion:
   - The x-axis shows decreasing amounts of regularization from left to right.
   - For strong regularization, the best solution is to have all values be exactly zero.
   - As the regularization becomes weaker,one by one, the values of the different coefficients first shoot up, then stabilize.
   - At some point, they all plateau as we are probably close to the unpenalized solution.
